```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('tokenized_access_logs.csv')

# Convert 'Date' column to datetime format and extract additional features
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %H:%M')
df['Day'] = df['Date'].dt.day
df['Year'] = df['Date'].dt.year

# Set up Seaborn visual style
sns.set(style='whitegrid', palette='pastel')

# Countplot of logs by month
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Month', data=df, order=df['Month'].value_counts().index)
plt.title('Log Distribution by Month')
plt.ylabel('Log Count')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()


```

```python
# Countplot of logs by hour
plt.figure(figsize=(10, 5))
sns.countplot(x='Hour', data=df, order=df['Hour'].value_counts().index)
plt.title
```

TOP 10 ACCESED PRODUCTS

```python
# Top 10 most accessed products
plt.figure(figsize=(10, 5))
top_products = df['Product'].value_counts().head(10)
sns.barplot(x=top_products.index, y=top_products.values)
plt.title('Top 10 Most Accessed Products')
plt.ylabel('Log Count')
plt.xlabel('Product')
plt.xticks(rotation=90)
plt.show()
```

 top 10 most frequent departments


```python
plt.figure(figsize=(10, 5))
top_departments = df['Department'].value_counts().head(10)
sns.barplot(x=top_departments.index, y=top_departments.values)
plt.title('Top 10 Most Accessed Departments')
plt.ylabel('Log Count')
plt.xlabel('Department')
plt.xticks(rotation=45)
plt.show()
```

top 10 most frequent categories

```python

plt.figure(figsize=(10, 5))
top_categories = df['Category'].value_counts().head(10)
sns.barplot(x=top_categories.index, y=top_categories.values)
plt.title('Top 10 Most Accessed Categories')
plt.ylabel('Log Count')
plt.xlabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(10, 5))
sns.countplot(x='Hour', data=df, order=df['Hour'].value_counts().index)
plt.title('Log Distribution by Hour of Day')
plt.ylabel('Log Count')
plt.xlabel('Hour')
plt.show()
```

logs by month

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='Month', data=df, order=df['Month'].value_counts().index)
plt.title('Log Distribution by Month')
plt.ylabel('Log Count')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.show()
```

descriptive statistics feautures

```python
# Count unique values for Category, Department, and Product
category_counts = df['Category'].value_counts()
department_counts = df['Department'].value_counts()
product_counts = df['Product'].value_counts()


print("Top 5 Categories:\n", category_counts.head(5))
print("\nTop 5 Departments:\n", department_counts.head(5))
print("\nTop 5 Products:\n", product_counts.head(5))

df.describe(include='all')
```

peak usage markdown

```python

monthly_usage = df.groupby('Month').size()
hourly_usage = df.groupby('Hour').size()


plt.figure(figsize=(10, 5))
monthly_usage.plot(kind='bar')
plt.title('Monthly Log Distribution')
plt.xlabel('Month')
plt.ylabel('Log Count')
plt.show()

plt.figure(figsize=(10, 5))
hourly_usage.plot(kind='bar')
plt.title('Hourly Log Distribution')
plt.xlabel('Hour of Day')
plt.ylabel('Log Count')
plt.show()
```

unique IP addresses (representing unique visitors)

```python

unique_visitors = df['ip'].nunique()
print(f"Number of unique visitors (unique IPs): {unique_visitors}")
```

Category and Department Correlation

```python
# the relationship between Categories and Departments
category_department_crosstab = pd.crosstab(df['Category'], df['Department'])


print(category_department_crosstab)


plt.figure(figsize=(10, 8))
sns.heatmap(category_department_crosstab, cmap="YlGnBu", annot=False)
plt.title('Category vs Department Correlation')
plt.xlabel('Department')
plt.ylabel('Category')
plt.show()
```

Time Series Analysis (Trends Over Time)

```python
# Group by Date to analyze log trends over time
daily_usage = df.groupby(df['Date'].dt.date).size()

# Plot daily log counts
plt.figure(figsize=(12, 6))
daily_usage.plot()
plt.title('Daily Log Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Log Count')
plt.grid(True)
plt.show()
```
